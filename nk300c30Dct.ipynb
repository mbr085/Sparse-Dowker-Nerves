```python
import numpy as np
import pandas as pd
from scipy.io import loadmat
from dowker_homology import dowker_persistent_homology, plot_persistence
%matplotlib notebook
```

```python
# read data
mat = loadmat("data/nk300c30Dct.mat")
coords = mat["nk300c30Dct"]
print(coords.shape)
```

```python
# choose parameters
n = 600
interleaving = 10
dimension = 1
```

```python
%%time 
homology_sheehy, cover_radius_sheehy, max_filtration_value_sheehy = dowker_persistent_homology(
    coords, 
    n_samples=n, 
    interleaving=interleaving,
    dimension=dimension, 
    method="sheehy",
    verbose=True)
```

```python
plot_sheehy = plot_persistence(homology_sheehy,
                               ticks=None, 
                               interleaving=interleaving,
                               cover_radius=cover_radius_sheehy,
                               max_filtration_value=max_filtration_value_sheehy,
                               title='nk3000c30, n=600, c=10',
                               s=10)
```
